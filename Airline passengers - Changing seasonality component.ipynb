```python
#sometimes the growth in trend or changes in seasonality can be either additive or multiplicative.
#That is to say that the actual rate of change is either linear and steady or the rate of change is itself changing throughout time.
```

```python
df.columns = ['ds','y']
df['ds'] = pd.to_datetime(df['ds'])
```

```python
m = Prophet()
m.fit(df)
future = m.make_future_dataframe(50,freq='MS')
forecast = m.predict(future)
fig = m.plot(forecast)
```

```python
fig = m.plot_components(forecast);
```
